```python
import pandas as pd
import numpy as np
from pycaret.classification import *
```

```python
model_pca = pd.read_csv("saved-data/all models/models_pca.csv")
```

```python
del model_pca['Binary Targ']
del model_pca['Target']
```

```python
numDict = {"normal": 0, "horizontal-misalignment": 1, "imbalance": 2, "overhang": 3, "underhang": 4, "vertical-misalignment": 5}
```

```python
setup_classification = setup(data = model_pca, target = "Num Targ", fix_imbalance = True,)
```

```python
# best_model_classification = compare_models()
```

```python
lr = create_model('lr')
tune_lr = tune_model(lr, optimize = 'auc')
```

```python
lightgbm = create_model('lightgbm')
tune_lightgbm = tune_model(lightgbm, optimize = 'auc')
```

```python
plot_model(tune_lr)
# plot_model(tune_lr, plot = 'confusion_matrix')
```

```python
comp0 = model_pca[model_pca['Num Targ'] == 0]
comp1 = model_pca[model_pca['Num Targ'] == 1]
comp2 = model_pca[model_pca['Num Targ'] == 2]
comp3 = model_pca[model_pca['Num Targ'] == 3]
comp4 = model_pca[model_pca['Num Targ'] == 4]
comp5 = model_pca[model_pca['Num Targ'] == 5]
```

```python
comp01 = pd.concat([comp0, comp1], axis=0)
comp02 = pd.concat([comp0, comp2], axis=0)
comp03 = pd.concat([comp0, comp3], axis=0)
comp04 = pd.concat([comp0, comp4], axis=0)
comp05 = pd.concat([comp0, comp5], axis=0)
```
