```python
import pandas as pd
import numpy as np
from pycaret.classification import *
```

```python
model_pca = pd.read_csv("saved-data/all models/models_pca.csv")
del model_pca['Target']
del model_pca['Num Targ']
```

```python
setup_classification = setup(data = model_pca, target = "Binary Targ", fix_imbalance = True)
```

```python
best_model_classification = compare_models()
```

```python
rf = create_model('rf')
tune_rf = tune_model(rf, optimize = 'auc')
```

```python
lightgbm = create_model('lightgbm')
tune_lightgbm = tune_model(lightgbm, optimize = 'auc')
```

```python
lr = create_model('lr')
tune_lr = tune_model(lr, optimize = 'auc')
```
